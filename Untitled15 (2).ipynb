#Customer churn prediction: Telecom Churn Dataset
Customer churn, also known as customer retention, customer turnover, or customer defection, is the loss of clients or customers.

Telephone service companies, Internet service providers, pay TV companies, insurance firms, and alarm monitoring services, often use customer attrition analysis and customer attrition rates as one of their key business metrics because the cost of retaining an existing customer is far less than acquiring a new one. Companies from these sectors often have customer service branches which attempt to win back defecting clients, because recovered long-term customers can be worth much more to a company than newly recruited clients.

Companies usually make a distinction between voluntary churn and involuntary churn. Voluntary churn occurs due to a decision by the customer to switch to another company or service provider, involuntary churn occurs due to circumstances such as a customer's relocation to a long-term care facility, death, or the relocation to a distant location. In most applications, involuntary reasons for churn are excluded from the analytical models. Analysts tend to concentrate on voluntary churn, because it typically occurs due to factors of the company-customer relationship which companies control, such as how billing interactions are handled or how after-sales help is provided.

predictive analytics use churn prediction models that predict customer churn by assessing their propensity of risk to churn. Since these models generate a small prioritized list of potential defectors, they are effective at focusing customer retention marketing programs on the subset of the customer base who are most vulnerable to churn.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, accuracy_score, precision_score, recall_score, f1_score, precision_recall_curve, roc_curve
from sklearn.preprocessing import StandardScaler, LabelEncoder
from imblearn.over_sampling import SMOTE
import shap
import plotly.graph_objects as go
import os
import joblib
```

```python
# Step 1: Load the datasets
# (Paper 1: Handling customer churn data)
df1 = pd.read_csv('churn-bigml-20.csv')
df2 = pd.read_csv('churn-bigml-80.csv')

# Combine the datasets
data = pd.concat([df1, df2], ignore_index=True)
```

Objective: Convert categorical data into numerical values for machine learning models.
Label Encoding: Each unique category is assigned a numeric code (e.g., Yes -> 1, No -> 0).

```python
# Preprocess data to handle non-numeric values
categorical_cols = data.select_dtypes(include=['object']).columns
for col in categorical_cols:
    data[col] = LabelEncoder().fit_transform(data[col])

# Step 2.1: Customer Churn in Data
# Pie chart for churn distribution
churn_counts = data['Churn'].value_counts()
fig = go.Figure(data=[go.Pie(labels=churn_counts.index, values=churn_counts.values, hole=0.4)])
fig.update_layout(title_text='Customer Churn Distribution')
fig.show()
```

```python
# Step 2.2: Variable Distributions
# Visualizing numeric columns
data.hist(bins=15, figsize=(15, 10), layout=(4, 5))
plt.tight_layout()
plt.show()
```

Look for skewness, multi-modal distributions, or other anomalies.
Features like Total day minutes or Monthly charges may have significant variance.

 Handle Outliers Using KNN


```python
from sklearn.neighbors import LocalOutlierFactor

# Step 1: Identify Binary and Non-Binary Features
binary_features = [col for col in data.columns if data[col].nunique() == 2]
non_binary_features = [col for col in data.columns if col not in binary_features]

print("Binary Features:", binary_features)
print("Non-Binary Features:", non_binary_features)

# Step 2: Apply KNN (Local Outlier Factor) on Non-Binary Features
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)  # Adjust contamination if needed
outlier_flags = lof.fit_predict(data[non_binary_features])

# The LOF assigns -1 to outliers and 1 to inliers
inliers = outlier_flags == 1

# Filter the dataset to remove outliers
data_cleaned = data[inliers]
print("Shape before removing outliers:", data.shape)
print("Shape after removing outliers:", data_cleaned.shape)
```

Local Outlier Factor (LOF):
Identifies outliers based on the distance between a data point and its neighbors.
Result: Rows flagged as outliers are removed from the dataset.
Analysis: Improves model performance by reducing noise and extreme values.

```python
# Step 3.2: Correlation Matrix (Before Feature Elimination)
plt.figure(figsize=(12, 8))
corr_matrix_before = data.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr_matrix_before, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix - Before Feature Elimination')
plt.show()
```

```python
# Step 3.3: Drop Highly Correlated Features
threshold = 0.9
upper_triangle = corr_matrix_before.where(np.triu(np.ones(corr_matrix_before.shape), k=1).astype(bool))
# Find features with correlation above the threshold
to_drop = [column for column in upper_triangle.columns if any(upper_triangle[column] > threshold)]
data = data.drop(columns=to_drop)
print(f"Dropped columns due to high correlation: {to_drop}")
```

```python
# Feature Selection using Random Forest Feature Importance
X = data_cleaned.drop(columns=['Churn'])
y = data_cleaned['Churn']

rf = RandomForestClassifier(random_state=42)
rf.fit(X, y)
feature_importances = pd.DataFrame(rf.feature_importances_, index=X.columns, columns=['Importance']).sort_values(by='Importance', ascending=False)
selected_features = feature_importances[feature_importances['Importance'] > 0.01].index
print(f"Selected features based on importance: {list(selected_features)}")
X = X[selected_features]
```

Random Forest: Identifies features with the highest predictive power.
Analysis:
Discards less relevant features, improving model efficiency.

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Model Implementations
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "KNN": KNeighborsClassifier(),
    "Random Forest": RandomForestClassifier(random_state=42),
    "SVM (RBF)": SVC(kernel='rbf', probability=True, random_state=42),
    "XGBoost": XGBClassifier(eval_metric='logloss', random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42)
}

model_results = {}
for model_name, model in models.items():
    print(f"Training {model_name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1] if hasattr(model, 'predict_proba') else None
    model_results[model_name] = {
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, average='binary'),
        "Recall": recall_score(y_test, y_pred, average='binary'),
        "F1-Score": f1_score(y_test, y_pred, average='binary'),
        "ROC AUC": roc_auc_score(y_test, y_prob) if y_prob is not None else None,
        "Confusion Matrix": confusion_matrix(y_test, y_pred)
    }
    # Save the trained model
    model_filename = f"models/{model_name.replace(' ', '_')}.pkl"
    joblib.dump(model, model_filename)
    print(f"{model_name} saved to {model_filename}")
    
```

Models: Trains multiple classifiers (e.g., Logistic Regression, Random Forest) to predict churn.
Evaluation Metrics:
Accuracy, Precision, Recall, F1-Score: Measure overall and class-specific performance.
ROC AUC: Evaluates model discrimination ability.
Confusion Matrix:
Highlights true positives, false positives, etc.


```python
# Step 5: Display All Metrics for Each Model
for model_name, metrics in model_results.items():
    print(f"\nMetrics for {model_name}:\n")
    for metric, value in metrics.items():
        if metric != "Confusion Matrix":
            print(f"{metric}: {value:.4f}")
    print("Confusion Matrix:")
    print(metrics["Confusion Matrix"])
```

```python
# Step 5.1: Model Performance Metrics Over Training Dataset
metrics_df = pd.DataFrame({
    name: [result["Accuracy"], result["Precision"], result["Recall"], result["F1-Score"], result["ROC AUC"]]
    for name, result in model_results.items()
}, index=["Accuracy", "Precision", "Recall", "F1-Score", "ROC AUC"]).T

metrics_df.plot(kind='bar', figsize=(12, 8))
plt.title('Model Performance Metrics (Training Dataset)')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.legend(title='Metric')
plt.show()
```

ROC Curve:

```python
# Precision-Recall Curves
plt.figure(figsize=(10, 8))
for model_name, model in models.items():
    if hasattr(model, 'predict_proba'):
        precision, recall, _ = precision_recall_curve(y_test, model.predict_proba(X_test)[:, 1])
        plt.plot(recall, precision, label=model_name)
plt.title('Precision-Recall Curves')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend(loc='lower left')
plt.show()
```
